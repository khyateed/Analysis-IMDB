```python
import requests
import sys
import numpy as np
import pandas as pd
import time
time.sleep(5)
page = requests.get("http://www.movie-wave.net/reviews-by-composer/")
page.status_code
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
soup.prettify()
```

```python
soup.find_all(class_='page-content')
```

```python
test = soup.find_all('p')
```

```python
titles = []
for item in test:
    titles.append(item.find_all('strong'))
```

```python
df = pd.DataFrame()
```

```python
titles
```

```python
lst = []
for i in range(0,334):
    lst.append(soup.find_all('p')[i].text.strip())
lst
```

```python
def music(lst):
    music = []
    for item in lst:
        music.append(item.split('\n'))
    
    music2 = []
    for item in music:
        for obj in item:
#             if obj.endswith('*') or obj.endswith('2'):
            music2.append(obj)
    
    return music2
```

```python
movie_list = music(lst)
```

```python
len(movie_list)
```

```python
df['full_string'] = movie_list
```

```python
rating = []
for item in movie_list:
    if item.endswith('*****'):
        rating.append(5)  
        item.strip('*****')
    elif item.endswith('**** 1/2'):
        rating.append(4.5)
    elif item.endswith(' ****'):
        rating.append(4)
    elif item.endswith(' *** 1/2'):
        rating.append(3.5)
    elif item.endswith(' ***'):
        rating.append(3)
    elif item.endswith(' ** 1/2'):
        rating.append(2.5)
    elif item.endswith(' **'):
        rating.append(2)
    elif item.endswith(' * 1/2'):
        rating.append(1.5)
    elif item.endswith(' *'):
        rating.append(1)
    else:
        rating.append('None') # need to switch this to null
    
```

```python
len(rating)
```

```python
df['rating'] = rating
```
